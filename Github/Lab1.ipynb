### Question 1
Design a python script (q1.py) to diagnose if a person is overweight or underweight, using weight (pound) and height (feet). Using the following formulas:
Standard Weight: sw = (height*30.48-105)/0.454
And identify the weight classification with the following standards:
- Underweight: weight<sw*0.9
- Normal Weight: sw*0.9<=weight<=sw*1.1
- Overweight: sw*1.1<weight<=sw*1.2
- Class I Obesity: sw*1.2<weight<=sw*1.3
- Class II Obesity: sw*1.3<weight<=sw*1.4
- Class III Obesity: sw*1.4<weight

Note: When a user finishing inputs, you need to output (a) sw value with scale=3 (For example, the number 123.45 has a precision of 5 and a scale of 2.) (b) his/her health status.

```python
#Overweight or Underweight
#Standard weight (SW)
def weightScale():
    try:
        height = float(input('What is your height:'))
        weight = float(input('What is your weight:'))
        sw = ((height*30.48)-105)/0.454
        if weight<(sw*0.9):
            Class = 'Underweight'
        elif (sw*0.9)<=weight<=(sw*1.1):
            Class = 'Normal'
        elif (sw*1.1)<weight<=(sw*1.2):
            Class = 'Overweight'
        elif (sw*1.2)<weight<=(sw*1.3):
            Class = 'Class I Obesity'
        elif (sw*1.3)<weight<=(sw*1.4):
            Class = 'Class II Obesity'
        elif (sw*1.4)<weight:
            Class = 'Class III Obesity'
        print('You are', Class, 'The Standard weight is', "%.3f" % sw)
    except ValueError:
        print('Invalid Entry')
weightScale()
```

### Question 2
Generate a random number between 1-100, and prompt a dialog for a user to make a guess. If it matches the random number, prompt a message (how many times user tried) to the user and stop the program, otherwise prompt the user a message with feedback about the comparison (greater than, or smaller than) and ask the user to make another guess.

```python
import random
def predict():
        r = random.randrange(1,101)
        count=0
        while True:
            try:
                userGuess = int(input('input a random number between 1 and 100:'))
                count = count +1
                if userGuess == r:
                    print('You guessed right!, you guessed correctly in', count, 'attempts')
                    break
                elif userGuess < r:
                    print('Your prediction is lower than the random value')
                else:
                    print('Your prediction is higher than the random value')
            except ValueError:
                print('Invalid Entry')
predict()
```

### Question 3
Write a script to convert dd:mm:ss degree (from users’ input) to decimal degree, and prompt the results to users. Note – please advise users the input format (20:30:40), otherwise you will have hard time to interpret the input.

```python
def convert():
    print('Convert your degree in angular form (dd:mm:ss) to decimal degree')
    try:
        degree = input('input a degree in this format dd:mm:ss:')
        List = degree.split(':')
        decimalDegree = int(List[0])+(int(List[1])/60)+(int(List[-1])/3600)
        print(degree, 'in decimal degree is', '%.4f' %decimalDegree)
    except:
        print('Invalid input')
convert()
```

#### Question 4
Solve the puzzle of 123456789 
* ' 7 1 6 '
* '+ 2 3 8'
* '-------'
* ' 9 5 4 '

*  ' a b c '
* '+ d e f '
*  '-------'
*  ' g h i '

* a-i are unique numbers from [1,9]

* Write a script to find all combinations which meet the above requirement. How many solutions do you find? Please print out all qualified solutions and also the total number of solutions you find.


```python
import itertools


counter = 0
for i in itertools.permutations('123456789'):
    s = ''.join(i)
    #print(s)
    tNum = int(s[0:3]) #[12]
    mNum = int(s[3:6]) 
    lNum = int(s[6:])
    if (tNum+mNum == lNum):
        counter+=1
        print(s)
print('The total number of qualified unique combination is', counter, 'as shown above')
```
